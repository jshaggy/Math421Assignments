### Titanic Plots



```{r titanic}
library(ggplot2)
ggplot(titanic) + geom_point(mapping = aes(x = Age,y = Fare, color = Sex))
```


This plot looks at Age and Fare to see if the amound spent on Fare is influenced by the Age of the passenger. Sex is represented by colors. 

```{r titanic2}
ggplot(titanic) + geom_bar(mapping = aes(x = Survived, fill = Sex), position = "dodge")
```

This bar chart looks at how many people survived, broken down using color to illustrate Sex

```{r titanic3}
ggplot(titanic) + geom_histogram(mapping = aes(x = SibSp, fill = Embarked), position = "dodge")
```

This histogram looks at the number of sibilings that each passenger had based on which port they embarked from.

```{r titanic4, warning=FALSE}

ggplot(titanic) + geom_point(mapping = aes(x = Age,y = Fare, color = Survived), position = "dodge")
```

This plot shows the fare that was paid by class and the color shows if the passenger survived or not.

```{r titanic5}
ggplot(data=titanic) + geom_boxplot(mapping = aes(x= Pclass, y=Fare ))

```



This is a boxplot to show the shape of the distribution of Fare and Pclass (Class of the passenger)


```{r titanic6}
ggplot(titanic) + geom_point(mapping = aes(x= Parch, y=Fare ))

```


This point graph shows the Fare paid by the number of parents/siblings on board


```{r titanic7}
ggplot(data =titanic, mapping = aes(x = Fare, y = SibSp)) + 
  geom_point(mapping = aes(color = Pclass)) + geom_smooth()

```

This point graph, with a curve, shows how much a passenger paid, along with how many siblings they have, with color showing their class
