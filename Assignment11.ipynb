



#10
```{r}

adult = read.csv("adult.csv")
df = read.csv("adult.csv", header = FALSE)
colnames(adult) <- c("Age","WorkClass","Fnlwgt","Education","EducationNum","MaritalStatus","Occupation", "Relationship", "Race", "Sex", "CapitalGain", "CaptialLoss", "HoursPerWeek", "NativeCountry","Salary")
adult[adult== " ?"]=NA

Mode <- function(x) {
     ux <- unique(x)
     ux[which.max(tabulate(match(x, ux)))]}

A10 = function(x){
  for (i in 1:ncol(x)){
    if (is.numeric(x[,i])){
     preProcess_missingdata_model <- preProcess(adult, method='knnImpute')
preProcess_missingdata_model

    }else if(is.factor(x[,i])){
      x[,i][is.na(x[,i])] = Mode(x[,i])
    }
  }
  return(x)
}

adult = A10(adult)

#Gets rid of ?
l = levels(adult$WorkClass)
l[-1]
c("Private",l[-1])

c = levels(adult$NativeCountry)
c[-1]
c("Nepal",c[-1])

g = levels(adult$Occupation)
g[-1]
c("Scientist",g[-1])

levels(adult$Education) = c("HighSchool","HighSchool","HighSchool","Elementary","MiddleSchool","MiddleSchool","HighSchool","UnderGrad","UnderGrad","UnderGrad","PostGrad","HighSchool","PostGrad","Elementary","UnderGrad","UnderGrad")
levels(adult$Occupation) = c("Other","Office","Military","Labor","Office","Labor","Service","Other","Service","Service","Office","Military","Office","Office","Labor")
levels(adult$NativeCountry) = c("Asia","Asia", "NorthAmerica","Asia","SouthAmerica","NorthAmerica","NorthAmerica","SouthAmerica","CentralAmerica","Europe","Europe","Europe","Europe","CentralAmerica","NorthAmerica","Europe", "CentralAmerica","Asia","Europe","Asia","Asia","Europe","Europe","NorthAmerica","Asia","Asia","NorthAmerica","CentralAmerica","Asia","SouthAmerica","Asia", "Europe", "Europe","NorthAmerica","Europe","Europe","Asia","Asia","Asia","NorthAmerica","NorthAmerica","Asia","Europe")

adult$Salary = factor(adult$Salary)
dummies_model <- dummyVars(Salary ~ ., data=adult)

trainData_mat <- predict(dummies_model, newdata = adult)
trainData <- data.frame(trainData_mat)
trainData$Salary <- adult$Salary

adult = trainData


preProcess_missingdata_model <- preProcess(adult, method= c("center", "scale"))

trainData <- predict(preProcess_missingdata_model, newdata = adult)

adult = trainData


names(adult)[51] = "target"
splitIndex <- createDataPartition(adult$target, p = .70, list = FALSE, times = 1)
set.seed(2018)
train <- adult[ splitIndex,]
test <- adult[-splitIndex,]

mytree <- rpart(target ~ ., data = train, method = "class")

fancyRpartPlot(mytree)

pred <- predict(mytree,test, type = "class")
cm = confusionMatrix(data = pred, reference = test$target, positive = " >50K")

cm$overall[1]
cm$byClass[11]


```
